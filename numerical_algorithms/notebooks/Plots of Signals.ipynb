```python
import os

import random
import numpy as np

from scipy import signal

import matplotlib
import matplotlib.pyplot as plt
plt.style.use("tableau-colorblind10")
matplotlib.rc('xtick', labelsize=18) 
matplotlib.rc('ytick', labelsize=18) 

%matplotlib inline
```

```python
np.random.seed(0)

N = 100 # no. of data points
noise = 0.15 # measurement noise

# Set Parameters
Tp = 2
Kp = 3

# State Space Model
A = -1.0 / Tp
B = Kp / Tp
C = 1.0
D = 0.0
sys = signal.StateSpace(A, B, C, D)
t_true, y_true = signal.step(sys, N=N)

# "take" measurements
y_meas = np.add(y_true, noise * np.random.randn(y_true.shape[0]))

# make the plot
fig = plt.figure(figsize=(10, 7))
plt.plot(t_true, y_true, "-", label="signal", linewidth=2)
plt.plot(t_true, y_meas, ".", label="measurement", markersize=12)
plt.title("Signal for First Order System", fontsize=22)
plt.legend(fontsize=20, loc="lower right")
plt.show()
```

```python
np.random.seed(0)

N = 100 # no. of data points
noise = 0.15 # measurement noise

# Set parameters
Kp = 3
Tp = 1
zeta = 0.15

# State Space Model
A = [[0.0, 1.0], [-1.0 / Tp ** 2, -2.0 * zeta / Tp]]
B = [[0.0], [Kp / Tp ** 2]]
C = [1.0, 0.0]
D = 0.0
sys = signal.StateSpace(A, B, C, D)
t_true, y_true = signal.step(sys, N=N) # the signal

# "take" measurements
y_meas = np.add(y_true, noise * np.random.randn(y_true.shape[0]))

# make the plot
fig = plt.figure(figsize=(10, 7))
plt.plot(t_true, y_true, "-", label="signal", linewidth=2)
plt.plot(t_true, y_meas, ".", label="measurement", markersize=12)
plt.title("Signal for Second Order System", fontsize=22)
plt.legend(fontsize=20, loc="lower right")
plt.show()
```

```python
np.random.seed(90)
N = 100

t_true = np.sort(np.random.randn(N, 1), axis=0).reshape(-1)
y_true = np.sin(t_true * 2 * np.pi / 3)
y_meas = y_true + 0.25 * np.random.randn(*y_true.shape)

# make the plot
fig = plt.figure(figsize=(10, 7))
plt.plot(t_true, y_true, "-", label="signal", linewidth=2)
plt.plot(t_true, y_meas, ".", label="measurement", markersize=12)
plt.title("Sinusoidal Signal", fontsize=22)
plt.legend(fontsize=20, loc="lower right")
plt.show()
```
